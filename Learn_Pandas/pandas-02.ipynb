### Scaling

#### Z Score scaling

```python
def z_score(x):
    return (x - np.mean(x))/np.std(x)
```

#### min max scaling

```python
def min_max_scaling(x):
    return (x - min(x))/(max(x) - min(x))
```

#### maximum absolute scaling

```python
def max_abs_scaling(x):
    return x/max(x)
```

#### robust scaling

```python
def robust_scaling(x):
    q2 = np.median(x)
    q3 = np.percentile(x, 75)
    q1 = np.percentile(x, 25)
    iqr = q3-q1
    return (x - q2)/iqr
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
budapest_zscale = df.loc[:, ['BUDAPEST']].apply(z_score)['BUDAPEST']
print(min(budapest_zscale), max(budapest_zscale))
```

```python
plt.boxplot(budapest_zscale)
plt.show()
```

```python
budapest_minmaxscale = df.loc[:, ['BUDAPEST']].apply(min_max_scaling)['BUDAPEST']
print(min(budapest_minmaxscale), max(budapest_minmaxscale))
```

```python
plt.boxplot(budapest_minmaxscale)
plt.show()
```

```python
budapest_robustscale = df.loc[:, ['BUDAPEST']].apply(robust_scaling)['BUDAPEST']
print(min(budapest_robustscale), max(budapest_robustscale))
```

```python
plt.boxplot(budapest_robustscale)
plt.show()
```
