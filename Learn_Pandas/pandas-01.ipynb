```python
## vectorization
def hypotenuse(a,b):
    return (a**2 + b**2)**0.5
```

```python
df = pd.DataFrame(np.random.randint(1,20,(3000,3)), columns=['A','B','C'])
df['C'] = np.nan
df.head()
```

```python
df['C'] = hypotenuse(df.A, df.B)
df.head()
```

```python
### Challenge
z = np.zeros([10,10], dtype=int)
z_df = pd.DataFrame(z)
z_df.head()
```

```python
%%time
# make all diagonals as 1
for i in range(len(z_df)):
    z_df.iloc[i,i] = i+1

z_df.head(10)    
```

```python
%%time
# make all diagonals as 1
for i in range(len(z_df)):
    z_df.iat[i,i] = i+1

z_df.head(10) 
```
