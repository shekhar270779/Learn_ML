### Search sorted

`Search sort returns the position at which a new number to be inserted into array so that new array remains sorted
`


```python
np.set_printoptions(threshold=2000)
```

```python
# create an array

np.random.seed(100)
arr = np.random.randint(1, 30, size=100)
arr.sort()
arr
```

```python
num = 6
pos = np.searchsorted(arr, num)
print(f"{num} should be inserted at index {pos} in array")
```

```python
# reconstruct array
arr = np.r_[arr[0:pos], num, arr[pos:]]
print(arr)
```

```python
%%timeit
np.searchsorted(arr, [20, 29, 37])
```

### Bisect module

Bisect module also provides similar functionality like searchsorted()

```python
import bisect
```

```python
np.random.seed(100)
arr = np.random.randint(1, 30, size=100)
arr.sort()
arr
```

```python
num=6
bisect.bisect(arr, num)
```

### Vectorization

```python
def is_prime(n):
    if n > 1:
        for i in range(2, n):
            if n %i == 0:
                return False
        else:
            return True
```

```python
for x in [10, 11, 13, 23, 17, 19, 29, 28]:
    print(f"Is {x} prime ? {is_prime(x)}")
```

```python
is_prime([10, 11, 13, 23, 17, 19, 29, 28])
```

```python
# We are going to vectorize this function
is_prime_v = np.vectorize(is_prime)
```

```python
is_prime_v([10, 11, 13, 23, 17, 19, 29, 28])
```

```python
is_prime_v(99)
```

```python
is_prime_v1 = np.vectorize(is_prime, otypes=[bool])
```

```python
is_prime_v1([10,11,12,14,15])
```

### Ufuncs in Numpy

- Ufuncs are functions that operate element by element on whole arrays.
- It supports broadcasting, typecasting
- Most common functions are median(), mean() etc are ufuncs
- They are written in C and linked to Python via numpy\'s ufunc facility


```python
def hyp(s1, s2):
    return np.sqrt(s1**2 + s2**2)
```

```python
# works on scaler
hyp(6,8)
```

```python
hyp([2,3,4,5,6], [2,3,3,7,8])
```

```python
# Making a ufunc
hyp_v = np.frompyfunc(hyp, nin=2, nout=1)
```

```python
hyp_v([2,3,4,5,6], [2,3,3,7,8])
```

```python
# Use vectorize function
hyp_vec = np.vectorize(hyp, otypes=[np.float16])
hyp_vec([2,3,4,5,6], [2,3,3,7,8])
```

```python
### Sinewave interpolation
np.set_printoptions(suppress=True)

x = np.linspace(0, 4*np.pi, 25)
y = np.sin(x)
```

```python
# plot
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(x, y, 'ro-')
plt.show()
```

```python
# interpolate
xvals = np.linspace(0, 4*np.pi, 100)
yinterp = np.interp(xvals, x, y)

plt.plot(x,y,'ro')
plt.plot(xvals, yinterp,'-x')
plt.show()
```

```python
### Interpolate with Scipy
from scipy import interpolate
```

```python
fx = interpolate.interp1d(x, y)
ynew = fx(xvals)

plt.plot(x,y,'ro')
plt.plot(xvals, ynew, '-x')
plt.show()
```

```python
fx = interpolate.interp1d(x, y, kind='cubic')
ynew = fx(xvals)

plt.plot(x,y,'ro')
plt.plot(xvals, ynew, '-x')
plt.show()
```
