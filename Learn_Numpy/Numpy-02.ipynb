### Post 02: Meshgrid

Meshgrid is a numpy function that creates 2d rectangular arrays from two 1d arrays.
<br>It provides all possible combinations of 1d array.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (10,7)
```

```python
x = np.arange(-3,4,1)
y = np.arange(-5,6,1)
print("X:")
print(x)
print("\nY:")
print(y)
```

```python
xm, ym = np.meshgrid(x,y)
```

```python
plt.plot(xm, ym, 'ro')
plt.show()
```

```python
plt.plot(xm, ym, 'r-')
plt.plot(xm.T, ym.T, 'b-')
plt.show()
```

### Concatenate & Split

```python
a = np.arange(12).reshape(3,4)
a
```

```python
b = a * 2
b
```

#### concatenate horizontally

```python
# hstack takes only one single argument, we put both arrays as a tuple or list
np.hstack((a, b))
```

```python
np.column_stack([a,b])
```

```python
np.concatenate([a,b], axis=1)
```

#### concatenate vertically

```python
np.vstack([a,b])
```

```python
np.row_stack([a,b])
```

```python
np.concatenate([a,b], axis=0)
```

#### Splitting array

```python
arr = np.concatenate([a,b])
arr
```

```python
# split vertically into two pieces (horizontal cut)
np.vsplit(arr,2)
```

```python
# split horizontally into two pieces (vertical cut)
np.hsplit(arr, 2)
```
